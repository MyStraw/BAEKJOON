**힌트:** `def get_price(self):`를 자식 클래스에 정의하고, `return self.price * (1 - self.discount_rate)`를 계산합니다.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def get_price(self):
        return self.price

class DiscountedProduct(Product):
    def __init__(self, name, price, discount_rate):
        super().__init__(name, price)
        self.discount_rate = discount_rate
    
    # 여기에 get_price 메서드를 재정의하세요

p_normal = Product("물", 1000)
p_discount = DiscountedProduct("콜라", 2000, 0.2)

print(f"물 가격: {p_normal.get_price()}")
# print(f"콜라 할인가: {p_discount.get_price()}") # 1600.0이 출력되어야 함
```

`Animal` → `Pet` → `Dog` 순서로 상속되는 클래스들을 만드세요. `super().__init__()`을 연쇄적으로 호출하여 `Dog` 객체가 `name`, `owner`, `breed` 속성을 모두 갖도록 만드세요.

**힌트:** `Dog`의 `__init__`은 `Pet`의 `__init__`을, `Pet`의 `__init__`은 `Animal`의 `__init__`을 호출해야 합니다.

```python
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"(Animal) 이름: {self.name}")

class Pet(Animal):
    def __init__(self, name, owner):
        # 여기에 Animal의 생성자 호출
        self.owner = owner
        print(f"(Pet) 주인: {self.owner}")

class Dog(Pet):
    def __init__(self, name, owner, breed):
        # 여기에 Pet의 생성자 호출
        self.breed = breed
        print(f"(Dog) 품종: {self.breed}")

d1 = Dog("초코", "김파고", "푸들")
print(f"이름: {d1.name}, 주인: {d1.owner}, 품종: {d1.breed}")
```
